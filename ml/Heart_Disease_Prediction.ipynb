```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
disease_df=pd.read_csv("/content/framingham.csv")
```

```python
disease_df.head(5)
```

```python
disease_df.drop(columns=['education'],inplace=True,axis=1)
```

```python
disease_df.rename(columns={'male':'Sex_male'},inplace=True)
```

```python
disease_df.head(5)
```

```python
disease_df.describe()
```

```python
disease_df.dropna(axis = 0, inplace = True)
disease_df
```

```python
disease_df.TenYearCHD.value_counts()
```

```python
X = np.asarray(disease_df[['age', 'Sex_male', 'cigsPerDay',
                           'totChol', 'sysBP', 'glucose']])
y = np.asarray(disease_df['TenYearCHD'])

X = preprocessing.StandardScaler().fit(X).transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size = 0.3, random_state = 4)

print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
X_train[5]
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy of the model is =',
      accuracy_score(y_test, y_pred))
```

```python
import numpy as np
weights = logreg.coef_      # array of coefficients
bias = logreg.intercept_

# Save to .npy (NumPy binary format)
np.save("weights.npy", weights)
np.save("bias.npy", bias)

# Save to text file
np.savetxt("weights.txt", weights)
with open("bias.txt", "w") as f:
    f.write(str(bias))
```
